```python
import os
import urllib.request
import json
import requests

client_id = "UCGMmwaM2R0nnw8kkWjq"
client_secret = "8BUnVV3Luq"
keyword = '산'
encText = urllib.parse.quote(keyword)
display = 100
page = 11
savePath = f'/Users/nam-eunji/Project/crawling/udigo/img/{keyword}/'  # 폴더 링크는 수정해주세요.

# 폴더 생성
try:
    if not os.path.exists(savePath):
        os.makedirs(savePath)
except OSError:
    print('Error: Creating directory. ' + savePath)


# 크롤링
file_n = 1
for i in range(1, page):
    print('check i : ', i)
    url = f"https://openapi.naver.com/v1/search/image?query={encText}&display={display}&start={i*100}"

    headers = {
        'X-Naver-Client-Id': 'UCGMmwaM2R0nnw8kkWjq',
        'X-Naver-Client-Secret': '8BUnVV3Luq'
    }
    req = requests.get(url, headers=headers)

    if req.status_code == 200:
        html = req.text
        html = json.loads(html)['items']
        for img in html:
            try:
                # 이미지링크 확인
                print(img['link'])

                # 저장 파일명 및 경로
                FileName = os.path.join(savePath, f'{file_n}.jpg')

                # 파일명 출력
                print('full name : {}'.format(FileName))

                # 이미지 다운로드 URL 요청
                urllib.request.urlretrieve(img['link'], FileName)
                file_n += 1

            except urllib.error.HTTPError as e:
                print('oh no... : ', e.reason)
                pass

            except urllib.error.URLError as e:
                print('oh no... : ', e.reason)
                pass

        # 다운로드 완료 시 출력
        print("--------download succeeded--------")

    else:
        print("Error Code:" + str(req.status_code))
```
